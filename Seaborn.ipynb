# Doppler effect

```python
fig, ax = plt.subplots(1,2, figsize= (10,5))

sns.rugplot(data=tips, x= "total_bill", ax= ax[0])
sns.histplot(data=tips, x= "total_bill", ax= ax[1], kde=True);
```

```python
fig, ax = plt.subplots(1,2, figsize= (10,5))

sns.rugplot(data=tips, x= "total_bill",height=0.5, ax= ax[0])
sns.kdeplot(data=tips, x= "total_bill", ax= ax[1]);
```

```python
fig, ax = plt.subplots(1,2, figsize= (10,5))

sns.rugplot(data=tips, x= "total_bill",height=0.1, ax= ax[0])
sns.histplot(data=tips, x= "total_bill", bins=50,kde=True, ax= ax[1]);
```

```python
sns.set(style= "dark")
sns.histplot(data=tips, x= "tip",kde=True)

plt.show()
```

# 2D KDE Graph (isohypse-like)

```python
sns.kdeplot(data=tips, x = "total_bill", y ="tip");
```

```python
sns.kdeplot(data=tips, x = "total_bill", y ="tip", color="red", shade=True);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

sns.kdeplot(data=tips, x = "total_bill", y ="tip", color="red", shade=True, ax=ax[1])
sns.scatterplot(data=tips, x = "total_bill", y ="tip", color="yellowgreen", ax=ax[0]);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

sns.kdeplot(data=tips, x = "total_bill", y ="tip", color="red", shade=True, ax=ax[0])
sns.scatterplot(data=tips, x = "total_bill", y ="tip", color="yellowgreen", ax=ax[1]);
```

```python
sns.displot(data= penguins, 
            x= "body_mass_g",
            bins=20,
            rug=True,
            kde=True);
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,5))

sns.kdeplot(data=tips, x = "total_bill", y ="tip", color="purple", shade=True, ax=ax[0,0])
sns.scatterplot(data=tips, x = "total_bill", y ="tip", color="yellowgreen", ax=ax[0,1])

sns.histplot(data=tips, x = "total_bill", color="orange", ax=ax[1,0])
sns.rugplot(data=tips, x = "total_bill",height=.5, color="blue", ax=ax[1,1])
plt.tight_layout()
plt.show()
```
